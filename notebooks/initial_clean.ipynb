### Remove tweets without a URL

```python
df=df[df.urls!="[]"]
```

```python
df = df.drop_duplicates(['urls'], keep="last")
```

```python
df
```

```python
def article_ktmpost(url):
    try:
        url = ast.literal_eval(url)[0]
        req = requests.get(url, headers)
        soup = BeautifulSoup(req.content, 'html.parser')
        result = soup.find("div", {"class": "col-sm-8"})
        return result.text
    except Exception as e:
        print (str(e))
        return 404
```

```python
df['article'] = df.urls.progress_apply(article_ktmpost)
```

```python
df.to_csv("KathmanduPost.csv", index=False)
```
